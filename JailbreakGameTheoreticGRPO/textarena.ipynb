```python
import textarena as ta

# Initialize agents, attacker (Player 0) and defender (Player 1)
agents = {
    0: ta.agents.OpenRouterAgent(model_name="pipihand01/QwQ-32B-Preview-abliterated-linear50"),   # Attacker
    1: ta.agents.OpenRouterAgent(model_name="Qwen/Qwen3-4B-Instruct-2507"),  # Defender
    2: ta.agents.OpenRouterAgent(model_name="Qwen/Qwen3-4B-Instruct-2507"),  # Dummy agents for unused player slots
    3: ta.agents.OpenRouterAgent(model_name="Qwen/Qwen3-4B-Instruct-2507"),  # Dummy agents for unused player slots
}

# Initialize Adversarial Taboo environment
env = ta.make(env_id="Taboo-v0")

# Optional: wrap for visualization
env = ta.wrappers.SimpleRenderWrapper(env=env) 

env.reset(num_players=len(agents))   # Should be 2 for Taboo

done = False
while not done:
    player_id, observation = env.get_observation()
    action = agents[player_id](observation)
    done, step_info = env.step(action=action)

rewards, game_info = env.close()
```
